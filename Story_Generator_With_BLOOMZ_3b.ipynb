```python
import keras_nlp

bloomz = keras_nlp.models.BloomCausalLM.from_preset('/root/.cache/kagglehub/models/keras/bloom/keras/bloomz_3b_multi/1')
bloomz.summary()
```

```python
# word list for choosing a random linking word

word_list = ['in addition','furthermore', 'also', 'and', 'as well as', 'because of', 'since', 'obviously', 'indeed', 'especially', 'similarly', 'however', 'nevertheless',
 'nonetheless', 'still', 'although', 'even though', 'though', 'but', 'yet', 'despite', 'in spite of', 'in contrast', 'in comparison', 'while', 'on the other hand',
 'for example', 'for instance', 'then' ,',']
```

```python
import numpy as np
from colorama import Fore, Style

def create_story_pipeline(word_list : np.array, first_prompt : str, max_generated_length : int = 256, story_length :int = 15):
    story = []

    for i in range(story_length):
        random_linking_word = np.random.default_rng().permutation(word_list)[0]
        if i == 0:
            story.append(first_prompt)
            generated_text = bloomz.generate(first_prompt + ' ' +  random_linking_word , max_length = max_generated_length)
            story.append(generated_text)
        else:
            generated_text = bloomz.generate(story[i-1] + ' ' + random_linking_word , max_length = max_generated_length)
            story.append(generated_text)

    return story



# for printing the story
def print_story(generated_story: np.array) -> None:
    print("Created Story with BLOOM")
    print("The first prompt is -> "  + Fore.BLUE + str.upper(generated_story[0]) + '\n'+ Style.RESET_ALL)
    print(Fore.RED + generated_story[-1] + Style.RESET_ALL)
```

```python
story_1 = create_story_pipeline(word_list, 'I used to smoke', 1024, 15)
print_story(story_1)
```

```python
story_2 = create_story_pipeline(word_list, 'I want to buy a red car', 1024, 15)
print_story(story_2)
```

```python
story_3 = create_story_pipeline(word_list, 'When I was a child in the USA', 1024, 20)
print_story(story_3)
```

```python
story_4 = create_story_pipeline(word_list, 'My favorite snack is ice cream,' , 1024, 20)
print_story(story_4)
```
